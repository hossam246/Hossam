```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'healthcare_dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Overview:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values[missing_values > 0])
```

```python
# Basic statistics
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Visualize distributions of numeric features
plt.figure(figsize=(12, 6))
sns.histplot(df.select_dtypes(include=['number']), bins=30, kde=True)
plt.title("Distribution of Numeric Features")
plt.show()
```

```python
# Check correlation between numeric features
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()

```

```python
# Check class distribution if diagnosis is a categorical target
if 'Diagnosis' in df.columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df['Diagnosis'])
    plt.title("Diagnosis Class Distribution")
    plt.xticks(rotation=45)
    plt.show()
```
